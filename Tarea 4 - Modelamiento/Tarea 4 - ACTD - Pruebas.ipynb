# Modelo Completo DF Arreglado

```python
import statsmodels.formula.api as smf

#feat_Limpio = dflimpio.drop(columns=["date","quarter","department","day","targeted_productivity","actual_productivity","idle_men"])

#X_Limpio = feat_Limpio
#y_Limpio_Actual = dflimpio['actual_productivity']

#X_train_Limpio, X_test, y_train_LA, y_test = train_test_split(X_Limpio, y_Limpio_Actual, random_state=1)

# agregar constante explíticamente
#X_train_Limpio = sm.add_constant(X_train_Limpio)

# regresión usando mínimos cuadrados ordinarios (ordinary least squares - OLS) 
#model_LA = sm.OLS(y_train_LA, X_train_Limpio).fit()

model_AAC = smf.ols(formula="actual_productivity ~ C(date)+C(quarter)+C(department)+C(quarter)+C(team)+smv+wip+over_time+incentive+idle_time+idle_men+no_of_style_change+no_of_workers", data=dfarreglado).fit()
# resumen de resultados
print(model_AAC.summary())


```

# Modelo Reducido DF Arreglado

```python
#model_AAR = smf.ols(formula="actual_productivity ~ C(department)+C(quarter)+C(team)+smv+over_time+incentive+no_of_workers", data=dfarreglado).fit()
model_AAR = smf.ols(formula="actual_productivity ~ C(department)+C(team)+smv+over_time+incentive+no_of_workers", data=dfarreglado).fit()

# resumen de resultados
print(model_AAR.summary())
```

# Calculo F-Parcial DF Arreglado

```python
import scipy.stats as stats

ssr_comp=model_AAC.ess
ssr_red=model_AAR.ess
gle_comp=model_AAC.df_resid
sse_comp=model_AAC.ssr
mse_comp=sse_comp/gle_comp
glr_comp=model_AAC.df_model
glr_red=model_AAR.df_model

F_parcial=((ssr_comp-ssr_red)/(glr_comp-glr_red))/mse_comp
p_valor = stats.f.cdf(F_parcial,(glr_comp-glr_red),gle_comp)

print(F_parcial)
print("Ep p-value es " +str(round(p_valor,4))+ " entonces se acepta la hipotesis nula y el modelo reducido es mejor")
```
